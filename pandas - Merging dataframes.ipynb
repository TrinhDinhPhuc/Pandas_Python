# We got four different ways to merge data

```python
# cach 1: Dung dataFrame Concat
pd.concat([movies,tags])
```

Nhung ma cach nay dung stacked vertically (xep chong theo chieu doc)nen no co NaN gop 2 bang lai bang cach them mot indicies 0,1,2,3,...

```python
# cach 2: dung concat kieu inner join
pd.concat([movies,tags],axis=1,join="inner")
```

cach nay thi tot hon roi, no se gop 2 bang lam mot, nhung con 1 yeu diem nua la phan movieID van bi lap lai 2 lan

```python
# cach 3, dung append()
movies.append(tags)
```

cach nay thi giong y het cai dau tien, dung concat, chi khac cai la ko can dung numpy

```python
# cach cuoi cung, merge inner
pd.merge(movies,tags,how="inner")
```

# Combine aggreagation, merging and filters to get useful analytics

```python
avg_ratings=ratings.groupby("movieId", as_index=False).mean()
```

```python
del avg_ratings["userId"]
```

```python
# loai bo cot userId  
```

```python
box_office= movies.merge(avg_ratings,on="movieId",how="inner")
box_office
```

sau do merge inner

```python
# apply filters
is_highly_rated= box_office["rating"]>=4.0
box_office[is_highly_rated][:5]
```

```python
# another filter
is_comedy = box_office["genres"].str.contains("Comedy")
box_office[is_comedy][:6]
```

#hay phet hehe

```python
# another filter use "&"
box_office[is_comedy & is_highly_rated][-5:]
```
