```python
import pandas as pd
```

```python
movies = pd.read_csv("ml-20m/movies.csv",sep=",")
```

```python
movies.head()
```

# Spilt 'genres' into multiple columns

```python
movie_genre= movies["genres"].str.split("|")
```

```python
movie_genre[:10]
```

```python
movie_genre=movies["genres"].str.split("|",expand=True)
movie_genre[:10]
```

# add a new column for comedy genre flag

```python
movie_genre['isComedy']=movies['genres'].str.contains('Comedy')
movie_genre
```

```python
# Extract year from title e.g (1995)
movies['year']=movies['title'].str.extract(".*\((.*)\).*",expand=True)
movies.tail(20)
```
