```python
Ca=[]
for i in data['Check in']:
    if (i>=datetime.time(hour=8, minute=0)and i<=datetime.time(hour=12, minute=0)):
        #print ('ca1')
        Ca.append('Ca Sáng')
    else:
        #print ('ca2')
        Ca.append('Ca Chiều')
```
